## Finding top 10 companies with the largest market cap

### Converting MarketCap to number

```python
def convfloat(strg):
    if strg == "n/a":
        return np.NaN
    else:
        strg1= strg.replace(strg[-1],"")
        strg2= strg1.replace("$","")
        value = float(strg2)
        if strg[-1] == "M":
            value = value * 1000000
        elif strg[-1] == "B":
            value = value * 1000000000
        return value

df["MarketCap"] = df["MarketCap"].apply(convfloat)
```

```python
df["MarketCap"].value_counts()
```

```python
df.dropna(subset=["MarketCap"],inplace=True)
```
